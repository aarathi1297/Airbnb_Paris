```python
stations = pd.read_pickle('airbnb_paris/stations_for_map.p')
```

```python
results = pd.read_pickle('airbnb_paris/recommendation_results_0615.p')
```

```python
paris_map = folium.Map(location=[stations['latitude'].mean(), 
                                 stations['longitude'].mean()], 
                      zoom_start=12)
#creating a Marker for each point in df_sample. Each point will get a popup with their zip
for row in stations.itertuples():
    paris_map.add_child(folium.Marker(location=[row.latitude,row.longitude],
           popup=row.name))
paris_map
```

```python
for row in results.itertuples():
    paris_map.add_child(folium.Marker(location=[float(row.latitude),float(row.longitude)],
           popup=row.id,icon=folium.Icon(color='red')))
paris_map
```

```python
df_attractions = pd.read_csv('airbnb_paris/Paris_attractions.csv')
```

```python
for i, row in df_attractions.iterrows():
    name = row['name'].replace("'",'')
    df_attractions.name.loc[i] = name
    
df_attractions
```

```python
for row in df_attractions.itertuples():
    paris_map.add_child(folium.Marker(location=[row.latitude,row.longitude],
           popup=row.name,icon=folium.Icon(color='green')))
paris_map
paris_map.save('paris_map.html')
```
